#### Cleaning and Importing Data

```python
import pandas as pd
import numpy as np
df = pd.read_csv('titanic1/full1.csv')
```

```python
# index
df.index
```

```python
# figure out data types
df.dtypes
```

```python
# Filtered all columns I didn't need
df.drop(['Ticket','Boarded', 'Destination','SibSp','Age','Age_wiki', 'Class', 'Parch','Cabin', 'WikiId', 'Embarked', 'Name_wiki', 'Lifeboat', 'Body'], axis=1)
```

#### Finding how many unique passengers there are

```python
len(set(df['Name']))
```

#### Finding the amount of First Class male passengers who survived

```python
# finding first class male passengers who survived
df.loc[df['Pclass']== 1]['Survived'][df['Sex']== 'male'].sum()
```

#### Finding how many Second or Third class pangers who survived

```python
# finding Second Class survival
list1 = df.loc[df['Pclass']== 2]['Survived'].sum()
list1
```

```python
# finding Third Class survival
list2 = df.loc[df['Pclass']== 3]['Survived'].sum()
list2
```

```python
# total number of Second and Third Class passengers who survived
list3 = list1 + list2
list3
```

#### Finding the survival rate of First Class Women 

```python
# finding survival rate of first class women
n = df.loc[df['Pclass']== 1]['Survived'][df['Sex']== 'female'].sum()
m = df.loc[df['Pclass']== 1]['Survived'][df['Sex']== 'female'].count()

survival_rate_firstclasswomen = n / m
survival_rate_firstclasswomen
```

#### Finding the survival rate of Third Class Men

```python
# finding survival rate of third class men
o = df.loc[df['Pclass']== 3]['Survived'][df['Sex']== 'male'].sum()
p = df.loc[df['Pclass']== 3]['Survived'][df['Sex']== 'male'].count()

survival_rate_thirdclassmen = o / p 
survival_rate_thirdclassmen
```

#### Finding the most common Hometown of passengers

```python
# most common hometown of passengers
df['Hometown'].mode()
```

#### Finding the Average Ticket Price

```python
# average ticket price
fare_avg = df.Fare.mean()
fare_avg
```
